```python
def swap(arr, i, j):
    arr[i], arr[j] = arr[j], arr[i]
```

```python
def partition(a, lo, hi):
    pivot = a[lo]
    i, j = lo + 1, hi

    while True:
        while i <= hi and a[i] < pivot:
            i += 1

        while j >= lo and a[j] > pivot:
            j -= 1

        if i >= j:
            break

        swap(a, i, j)
    swap(a, lo, j)
    return j
```

```python
def sort(a, lo, hi):
    if hi <= lo:
        return

    j = partition(a, lo, hi)
    sort(a, lo, j - 1)
    sort(a, j + 1, hi)
```

```python
import random

if __name__ == "__main__":
    a = [i for i in range(1000)]
    random.shuffle(a)
    print(a)
    partition(a, 0, len(a) - 1)
    sort(a, 0, len(a) - 1)
    print(a)
    print(a == sorted(a))
```
